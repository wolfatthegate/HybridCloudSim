```python
import random

NUMOFJOBS = 20
arrival_time = 0
jobs = {}

for i in range(0, NUMOFJOBS):
    job_id = i
    num_qubits = random.randint(3, 15)
    depth = random.randint(5, 20)
    priority = 2
    arrival_time += random.expovariate(lambd=1.5)
    num_shots = random.randint(5, 20) * 100
    req_iterations = random.randint(3, 7)

    # CPU demand scales with qubit count (2–10)
    cpu_units = int(
        max(2, min(10, round(2 + (num_qubits / 15) * 8)))
    )

    # Memory bandwidth demand scales with qubit count (3–20)
    mem_bw = int(
        max(3, min(20, round(3 + (num_qubits / 15) * 17)))
    )

    print(
        job_id,
        num_qubits,
        depth,
        priority,
        round(arrival_time, 4),
        num_shots,
        req_iterations,
        cpu_units,
        mem_bw
    )
```

```python
import random
import csv

NUMOFJOBS = 2000
arrival_time = 0.0

filename = "2000-job-2.csv"

with open(filename, mode="w", newline="") as f:
    writer = csv.writer(f)

    # Write header
    writer.writerow([
        "job_id",
        "num_qubits",
        "depth",
        "priority",
        "arrival_time",
        "num_shots",
        "req_iterations", 
        "cpu_units", 
        "mem_bw"
    ])

    # Generate jobs
    for i in range(NUMOFJOBS):
        job_id = i
        num_qubits = random.randint(3, 15)
        depth = random.randint(5, 20)
        priority = 2
        arrival_time += random.expovariate(lambd=1.5)
        num_shots = random.randint(5, 20) * 100
        req_iterations = random.randint(3, 7)
        # CPU demand scales with qubit count (2–10)
        cpu_units = int(
            max(2, min(10, round(2 + (num_qubits / 15) * 8)))
        )

        # Memory bandwidth demand scales with qubit count (3–20)
        mem_bw = int(
            max(3, min(20, round(3 + (num_qubits / 15) * 17)))
        )
        writer.writerow([
            job_id,
            num_qubits,
            depth,
            priority,
            round(arrival_time, 4),
            num_shots,
            req_iterations, 
            cpu_units, 
            mem_bw
        ])

print(f"Wrote {NUMOFJOBS} jobs to {filename}")
```
